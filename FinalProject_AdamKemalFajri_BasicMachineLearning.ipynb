### Dara Preprocessing

```python
#dilakukan proses label encoder untuk fitur diagnosis
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['diagnosis'] = le.fit_transform(df['diagnosis'])

df['diagnosis'].value_counts()
```

```python
#dilakukan penghapusan fitur id dan Unnamed: 32 pada dataframe karena tidak diperlukan dalam pembuatan model
df.drop(['id','Unnamed: 32'], axis = 1, inplace = True)
```

```python
#dataframe baru setalah dilakukan LabelEncoder dan drop fitur id dan Unnamed: 32
df
```

```python
#dilakukan proses pemisahan data untuk feature dan target
x = df.drop(['diagnosis'],axis=1) #feature
y = df['diagnosis'] #target
print('dimensi x :', x.shape)
print('dimensi y :', y.shape)
```

```python
#dilakukan proses pemisahan data untuk data training (75%) dan data test(25%)
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 16)
```

```python
#melihat dimensi dari data
print(x_train.shape)
print(x_test.shape)
```

```python
#dilakukan proses penstandaran data untuk train dan test
from sklearn.preprocessing import StandardScaler

stc = StandardScaler()

x_train = stc.fit_transform(x_train)
x_test = stc.fit_transform(x_test)
```

### Data Modeling

```python
'''Pada pembuatan model ini saya menggunakan confusion_matrix yang mana,
Berikut adalah beberapa manfaat dari confusion matrix:
1.Menunjukkan bagaimana model ketika membuat prediksi.
2.Tidak hanya memberi informasi tentang kesalahan yang dibuat oleh model tetapi juga jenis kesalahan yang dibuat.
3.Setiap kolom dari confusion matrix merepresentasikan instance dari kelas prediksi.
4.Setiap baris dari confusion matrix mewakili instance dari kelas aktual.'''
```

#### Support Vector Machine

```python
#import library untuk SVC dan library lainnya yg dibutuhkan
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import GridSearchCV, cross_val_score
```

```python
#pembuatan model SVC
model = SVC()

model.fit(x_train, y_train)
y_pred = model.predict(x_test)

print("Training accuracy :", model.score(x_train, y_train))
print("Testing accuarcy :", model.score(x_test, y_test))
svc = model.score(x_test, y_test)
cr = classification_report(y_test, y_pred)
print(cr)

cm1 = confusion_matrix(y_test, y_pred)
print(cm1)
plt.rcParams['figure.figsize'] = (5, 5)
sns.heatmap(cm1, annot = True, cmap = 'winter')
plt.title('Confusion Matrix', fontsize = 20)
plt.show()
```

```python
# pencarian nilai/besaran parameter terbaik untuk model SVC
param = {
    'C': [0.8,0.9,1,1.1,1.2,1.3,1.4],
    'kernel':['linear', 'rbf','poly','sigmoid'],
    'gamma' :[0.1,0.8,0.9,1,1.1,1.2,1.3,1.4]
}
#digunakan GridCV
grid_svc = GridSearchCV(model, param_grid = param, scoring = 'accuracy', cv = 10)
```

```python
grid_svc.fit(x_train, y_train)
```

```python
#diperoleh nilai parameter terbaik
grid_svc.best_params_
```

```python
#pembuatan model baru SVC dengan memasukan nilai parameter yang telah dicari tersebut
model2 = SVC(C = 0.8, gamma = 0.1, kernel = 'linear')
model2.fit(x_train, y_train)
y_pred = model2.predict(x_test)

print(classification_report(y_test, y_pred))

print("Testing accuarcy :", model2.score(x_test, y_test))
svc2 = model2.score(x_test, y_test)
```

#### Logistic Regression

```python
#import library LogisticRegression
from sklearn.linear_model import LogisticRegression

#pembuatan model LogisticRegression
lr = LogisticRegression()

lr.fit(x_train, y_train)

y_pred = lr.predict(x_test)

print("Training accuracy :", lr.score(x_train, y_train))
print("Testing accuarcy :", lr.score(x_test, y_test))
logreg = lr.score(x_test, y_test)
cr = classification_report(y_test, y_pred)
print(cr)
 
cm2 = confusion_matrix(y_test, y_pred)
print(cm2)
plt.rcParams['figure.figsize'] = (5, 5)
sns.heatmap(cm2, annot = True, cmap = 'winter')
plt.title('Confusion Matrix', fontsize = 20)
plt.show()
```

```python
#pencarian nilai parameter terbaik menggunakan gridcv
LR_params = {'C':[0.001, 0.1, 1, 10, 100]}
grid_lr = GridSearchCV(lr, param_grid = LR_params, scoring = 'accuracy', cv = 10)
```

```python
grid_lr.fit(x_train, y_train)
```

```python
#diperoleh nilai parameter terbaik
grid_lr.best_params_
```

```python
#pembuatan model LogisticRegression baru dengan memasukan nilai parameter yang baru tadi
lr2 = LogisticRegression(C = 1)
lr2.fit(x_train, y_train)
y_pred = lr2.predict(x_test)

print(classification_report(y_test, y_pred))

print("Testing accuarcy :", lr2.score(x_test, y_test))
logreg2 = lr2.score(x_test, y_test)
```

```python
from sklearn import metrics
y_pred_proba = lr.predict_proba(x_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred)
auc = metrics.roc_auc_score(y_test, y_pred)
plt.plot(fpr,tpr,label="data 1, auc="+str(auc))
plt.legend(loc=4)
plt.show()
```

#### Random Forest

```python
#import library RandomForestClassifier
from sklearn.ensemble import RandomForestClassifier

#pembuatan model RandomForestClassifier
rf = RandomForestClassifier(n_estimators = 400, max_depth = 10)

rf.fit(x_train, y_train)
y_pred = rf.predict(x_test)

print("Training accuracy :", rf.score(x_train, y_train))
print("Testing accuarcy :", rf.score(x_test, y_test))
ranfor = rf.score(x_test, y_test)
cr = classification_report(y_test, y_pred)
print(cr)

cm3 = confusion_matrix(y_test, y_pred)
print(cm3)
plt.rcParams['figure.figsize'] = (5, 5)
sns.heatmap(cm3, annot = True, cmap = 'winter')
plt.title('Confusion Matrix', fontsize = 20)
plt.show()
```

```python
#pencarian nilai parameter terbaik
RF_params = {'n_estimators':[10,50,100,150,200],'max_depth':[5,10,15,20,25,30,35,40]}
grid_rf = GridSearchCV(rf, param_grid = RF_params, scoring = 'accuracy', cv = 10)
```

```python
grid_rf.fit(x_train, y_train)
```

```python
#diperoleh nilai parameter terbaik dari gridCV
grid_rf.best_params_
```

```python
#pembuatan model RandomForestClassifier baru dengan memasukan nilai parameter baru
rf2 = RandomForestClassifier(max_depth= 40, n_estimators= 1000)
rf2.fit(x_train, y_train)
y_pred = rf2.predict(x_test)

print(classification_report(y_test, y_pred))

print("Testing accuarcy :", rf2.score(x_test, y_test))
ranfor2 =  rf2.score(x_test, y_test)
```

Pada model diatas, ketika dilakukan penambahan n_estimators yang divariasikan semakin tinggi dan max_depth tetap sebesar 40 maka akan menghasilkan hasil akurasi score yang lbih baik dari model defaultnya

#### Decision Tree

```python
#import library DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier

#pembuatan model DecisionTreeClassifier
dfr = DecisionTreeClassifier()

dfr.fit(x_train, y_train)
y_pred = dfr.predict(x_test)

print("Training accuracy :", dfr.score(x_train, y_train))
print("Testing accuarcy :", dfr.score(x_test, y_test))
decitree = dfr.score(x_test, y_test)
cr = classification_report(y_test, y_pred)
print(cr)
 
cm4 = confusion_matrix(y_test, y_pred)
print(cm4)
plt.rcParams['figure.figsize'] = (5, 5)
sns.heatmap(cm4, annot = True, cmap ="Blues" )
plt.title('Confusion Matrix', fontsize = 20)
plt.show()
```

```python
#pembuatan model baru DecisionTreeClassifier dengan menambahkan nilai parameter yang diperoleh secara random/tebakan
dfr2 = DecisionTreeClassifier(criterion = 'gini', max_depth= 110)
dfr2.fit(x_train, y_train)
y_pred = dfr2.predict(x_test)

print(classification_report(y_test, y_pred))

print("Testing accuarcy :", dfr2.score(x_test, y_test))
decitree2 = dfr2.score(x_test, y_test)
```

pada model diatas, ketika dilakukan variasi max_depth diperbesar mendekati 100 maka akurasi model akan meningkat, namun ketika max_depth diperbedar mendekati 1000 maka akurasinya akan menurun dibanding ketika max_depth diset sebesar 100
